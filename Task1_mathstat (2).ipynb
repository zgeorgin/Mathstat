```python
import numpy as np
import matplotlib.pyplot as plt
import plotly
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots # Подключаем нужные библиотеки
```

```python
theta = 5
sample = np.sort(np.random.uniform(0, theta, 100)) # Объявляем параметр и  генерируем выборку
```

```python
def construct_array(theta_function, sample): # Вспомогательная функция для построения точек
    current_sample = []
    theta_array = []
    for i in sample:
        current_sample.append(i)
        theta_array.append(theta_function(current_sample))
    return np.array(theta_array)
```

```python
def theta1(x): # Функция для 1 оценки
    return 2 * np.mean(x)
```

```python
def theta3(x): # функция для 3 оценки
    return ((len(x) + 1)/len(x)) * np.max(x)
```

```python
def theta4(x): # функция для 4 оценки
    return np.max(x) + np.min(x)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=np.arange(1,101), y=abs(construct_array(theta1, sample) - theta), name = "1 оценка"))
fig.add_trace(go.Scatter(x=np.arange(1,101), y=abs(construct_array(theta3, sample) - theta), name = "3 оценка"))
fig.add_trace(go.Scatter(x=np.arange(1,101), y=abs(construct_array(theta4, sample) - theta), name = "4 оценка"))
```
